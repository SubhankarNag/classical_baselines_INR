```python
import matplotlib.pyplot as plt
import numpy as np

codecs = {
    "JPEG": jpeg,
    "JPEG2000": jpeg2000,
    "spiht":spiht,
    "H264": h264,
    "H265": h265,
    "ICAINR": icainr,
    "Ours": deadnf,
}
v_r = ["Volume", "RSOS"]
for i in range(1, 3):
    plt.figure(figsize=(10, 5))

    for name, data in codecs.items():
        arr = np.array(data)
        # print(name, arr.shape)
        ratio = arr[:, 0]
        vol_rrmse = arr[:, i]

        # Sort by ratio for a cleaner curve
        idx = np.argsort(ratio)

        plt.plot(
            ratio[idx],
            vol_rrmse[idx],
            marker="o",
            label=name
        )

    plt.xlabel("Compression Ratio", fontsize=16)
    plt.ylabel(f"{v_r[i-1]} RRMSE", fontsize=16)
    plt.title("RD Curves", fontsize=16)
    plt.grid(True)
    plt.legend(fontsize=14)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.tight_layout()
    plt.savefig(f"rd_curve_{v_r[i-1]}.png", dpi=300, bbox_inches="tight")
    plt.show()
    plt.close()
```
